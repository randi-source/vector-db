```python
# Setup a client
import chromadb
client = chromadb.Client()
```

```python
# Create a 'collection'. Think of it like a 'table' in a vector database
collection1 = client.create_collection(name="my_collection1")
```

```python
collection2 = client.create_collection(name="my_collection2")
```

```python
print(collection1)
```

```python
collection1.add(
    documents=["This is a document", "This is another document"],
    metadatas=[{"source":"source1", "language":"en"}, {"source":"source2"}],
    ids = ["id1", "id2"]
)
```

```python
# Methods can be run on clients
```

```python
client.list_collections()
```

```python
# Get an existing collection

collection1 = client.get_collection("my_collection1")
```

```python
# Get a collection or create if it doesnt exist already

collection3 = client.get_or_create_collection("my_collection3")
```

```python
# Delete a collection

client.delete_collection("my_collection3")
```

```python
# Utility methods

# heartbeat
client.heartbeat()
```

```python
# Methods on Collections
```

```python
# count number of items in a collection

collection1.count()
```

```python
# Get items from a collection
collection1.get()
```

```python
# Adding new items to collection

# One at a time

collection1.add(
    documents=["This is document 3", "This is document 4"],
    metadatas=[{"source":"source3", "language":"en"}, {"source":"source4"}],
    ids = ["id3", "id4"]
)
```

```python
collection1.get()
```

```python
# Upsert

collection1.upsert(
    documents=["doc3", "This is document 4"],
    metadatas=[{"source":"source3", "language":"en"}, {"source":"source4"}],
    ids = ["id3", "id4"]
)
```

```python
collection1.get()
```

```python
collection1.peek()
```
